```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.set_option('display.max_column',None)
pd.set_option('display.max_row',None)
```

```python
train = pd.read_csv("train.csv")
train.head()
```

```python
train.info()
```


#### In data analysis We Will analyse to find 
1. Missing Values
2. All the numerical values
3. Distribution of numerical values
4. Categorical Values
5. Outliers
6. Relationship b/w independent and dependent features

```python
# Missing values
train.isnull().sum().sort_values(ascending=False)
```

```python
feature_na = [features for features in train.columns if train[features].isnull().sum()>1]

for feature in feature_na:
    print(feature,np.round(train[feature].isnull().mean(),4), '%missing')    
```

### Now we will chwck whether the missing values impact our sales price or not

```python
for feature in feature_na:
    data=train.copy()
    data[feature] = np.where(data[feature].isnull(),1,0) #convert the feature which are null to 1 and rest to 0
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

The relation b/w the null values and sales price is visible and it does affect the price of the houses so we'll replace it with something meaningful

### Numerical Variables

```python
num_feature = [feature for feature in train.columns if train[feature].dtype!='O']
print('Number of numerical variables: ',len(num_feature))
train[num_feature].head()
```

### Temporal variables(E.g Data and time variables)


```python
# variables having year info
year_feature = [feature for feature in num_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
# analyse temporal variables
# let us find out if there is a r/s b/w yearsold and sale price

train.groupby('YrSold')['SalePrice'].median().plot()
plt.title('House Price vs year Sold')
```

```python
# comparing differences b/w all year feature with sale price

for feature in year_feature:
    if feature !='YrSold':
        data=train.copy()
        data[feature] = data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.show()
```

### Discrete and continuos variables

```python
discrete_feature = [feature for feature in num_feature if len(train[feature].unique())<25 and feature not in year_feature+['Id']]
print('Discrete variable: {}'.format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
# find the r/s between these discrete features and SalePrice
for feature in discrete_feature:
    data=train.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.title(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
# Continuous variables
cont_var = [feature for feature in num_feature if feature not in discrete_feature+year_feature+['Id']]
print('Continuous feature {}'.format(len(cont_var)))
```

```python
for feature in cont_var:
    data=train.copy()
    data[feature].hist(bins=20)
    plt.xlabel(feature)
    plt.title(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
# We will be using logarithmic transformation for dealing with skewed data


for feature in cont_var:
    data = train.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

# Outliers

```python
for feature in cont_var:
    data=train.copy()
    data.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```

##  Categorical Variables

```python
cat_features = [feature for feature in train.columns if data[feature].dtypes=='O']
```

```python
train[cat_features].head()
```

```python
for feature in cat_features:
    print('The feature is {} and the number of categories are {}'.format(feature,len(train[feature].unique())))
```

```python
## find the r/s bet categorical feature and saleprice

for feature in cat_features:
    data=train.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

## Missing Values

```python
feature_nan = [feature for feature in train.columns if train[feature].isnull().sum()>1 and train[feature].dtypes=='O']

for feature in feature_nan:
    print('{}: {}% missing values'.format(feature,np.round(train[feature].isnull().mean(),4)))
```

```python
sns.heatmap(train.isnull(),yticklabels=False)
```

```python
# Replacing categorical missing values
def replace_cat_feature(train,feature_nan):
        data=train.copy()
        data[feature_nan]=data[feature_nan].fillna('Missing')
        return data
train = replace_cat_feature(train,feature_nan)
sns.heatmap(train.isnull(),yticklabels=False)
```

```python
train.head()
```

```python
# Replacing numerical null values
numerical_nan = [feature for feature in train.columns if train[feature].isnull().sum()>1 and train[feature].dtypes!='O']

for feature in numerical_nan:
    print('{}: {}% missing value'.format(feature,np.around(train[feature].isnull().mean(),
                                                         4)))
```

```python
# replacing null numerica values with median values

for feature in numerical_nan:
    median_value=train[feature].median()
    train[feature+'NaN']=np.where(train[feature].isnull(),1,0)
    train[feature].fillna(median_value,inplace=True)
    
sns.heatmap(train.isnull(),yticklabels=False)
```

```python
# Temporal Variables

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    train[feature]=train['YrSold']-train[feature]
train['YrSold'].head()
```

```python
train[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
# transforming skewed data into log transformation
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for feature in num_features:
    train[feature] = np.log(train[feature])
```

# Handling Rare categorical features
the categorical features which are less than 1% are declared as rare

```python
for feature in cat_features:
    temp = train.groupby(feature)['SalePrice'].count()/len(train)
    temp_df = temp[temp>0.01].index
    train[feature] = np.where(train[feature].isin(temp_df),train[feature],'Rare_var')
```

```python
train.head(100)
```

## Feature Scaling

```python

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for column_name in train.columns:
    if train[column_name].dtype == object:
            train[column_name] = le.fit_transform(train[column_name])
    else:
            pass
```

```python
feature_scale = [feature for feature in train.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit_transform(train[feature_scale])
```

```python
data = pd.concat([train[['Id','SalePrice']].reset_index(drop=True),
                 pd.DataFrame(scaler.transform(train[feature_scale]),columns=feature_scale)],
                axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_Train.csv',index=False)
```

# Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
dataset = pd.read_csv('X_Train.csv')
dataset.head()
```

```python
x_train = dataset.drop(['SalePrice','Id'],axis=1)
y_train = dataset['SalePrice']
```

```python
# the lasso regression is used for selecting the features and the alpha
# value is set accordingly higher the value lesser is the features selected
# then selectfrommodel is used which will select features whose
# coefficients are non-zero

feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(x_train,y_train)
```

```python
# this will return values true or false for the features which are important
# to select and which are not respectively

feature_sel_model.get_support()
```

```python
selected_feat = x_train.columns[(feature_sel_model.get_support())]

print('total feature: {}'.format((x_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
```

```python
selected_feat
```

```python
x_train=x_train[selected_feat]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_train,y_train,test_size=0.20,random_state=42)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
ln = LinearRegression()
ln.fit(x_train,y_train)
```

```python
pred1 = ln.predict(x_test) 
```

```python
from sklearn.metrics import r2_score
R = r2_score(pred1,y_test)
print(R) #R-squared
```

```python
from sklearn import metrics

print(metrics.mean_absolute_error(y_test,pred1)) #MAE
print(metrics.mean_squared_error(y_test,pred1))  #MSE
print(np.sqrt(metrics.mean_squared_error(y_test,pred1))) #RMSE
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor 
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(x_train, y_train)
```

```python
pred2 = regressor.predict(x_test)
R2 = r2_score(pred2,y_test)
print(R2)
```

```python
mse = metrics.mean_squared_error(y_test, pred2)
rmse = mse**.5
print(mse)
print(rmse)
```

# XgBoost Regressor

```python
params = {
    "n_estimators": 500,
    "max_depth": 4,
    "learning_rate": 0.01,
    
}
```

```python
from sklearn import ensemble
reg = ensemble.GradientBoostingRegressor(**params)
reg.fit(x_train, y_train)
```

```python
pred3 = reg.predict(x_test)
R3 = r2_score(pred3,y_test)
print(R3)
```

```python
mse = metrics.mean_squared_error(y_test, pred3)
rmse = mse**.5
print(mse)
print(rmse)
```
