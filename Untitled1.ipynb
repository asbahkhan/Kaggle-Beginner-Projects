```python
def age_cal(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 38
        elif Pclass == 2:
            return 29
        else:
            return 25
    else:
        return Age
```

```python
train['Age'] = train[['Age','Pclass']].apply(age_cal,axis=1)
sns.heatmap(train.isnull(),yticklabels=False)
```

```python
train.drop('Cabin',axis=1,inplace=True)
```

```python
train.head()
```

```python
# Dummy variables
```

```python
Embark = pd.get_dummies(train['Embarked'],drop_first=True)
Sex = pd.get_dummies(train['Sex'],drop_first=True)
```

```python
train.drop(['Sex','Name','Ticket','Embarked'],axis=1,inplace=True)
```

```python
train = pd.concat([train,Sex,Embark],axis=1)
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train, y_test = train_test_split(train.drop('Survived',axis=1),train['Survived'],test_size=0.30,random_state=12)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lm = LogisticRegression()
lm.fit(x_train,y_train)
```

```python
pred = lm.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy = confusion_matrix(y_test,pred)
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
ac_score = accuracy_score(y_test,pred)
ac_score
```
